```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
data_arr = np.loadtxt('heightWeightData.txt', delimiter=',')
```

```python
male_data = data_arr[data_arr[:, 0]==1]
male_mean = male_data[:,1:].mean(axis=0)
male_cov = np.cov(male_data[:, 1:].transpose())
male_cov_det = np.linalg.det(male_cov)
male_cov_inv = np.linalg.inv(male_cov)
p_x_male = lambda x:(1/math.sqrt((2*math.pi)**2*male_cov_det))*np.exp((-1/2)*(x-male_mean).dot(male_cov_inv).dot((x-male_mean).transpose()))
```
